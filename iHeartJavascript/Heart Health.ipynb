```python
import pandas as pd
from sqlalchemy import create_engine, inspect
from sqlalchemy.orm import Session
import os
import sqlite3
import sqlalchemy

from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
heart_data = "static/heart.csv"
data_table = pd.read_csv(heart_data, header=None)
```

```python
heart_df = pd.read_csv(heart_data, encoding="ISO-8859-1")
```

```python
heart_df
```

```python
heart_df1 = heart_df.rename(columns={"":"id"})
heart_df1 = heart_df1.reset_index()
heart_df1 = heart_df1.rename(columns={"index": "id"})
heart_df1.set_index("id", inplace=True) 
heart_df1
```

```python
#This creates the sqlite database and saves it in the current directory
engine = create_engine("sqlite:///heart_database.sqlite")
```

```python
con = sqlite3.connect("heart_database.sqlite")
cur = con.cursor()
cur.execute('''DROP TABLE IF EXISTS participants;''')
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
heart_df1.to_sql(name='participants', con=engine, if_exists='append', index=True)
```

```python
list1 = {
    'men': {
        'asymptomatic': 0, 
        'atypical angina': 0, 
        'non-anginal pain': 0,
        'typical angina': 0
    }, 
    'women': {
        'asymptomatic': 0, 
        'atypical angina': 0, 
        'non-anginal pain': 0,
        'typical angina': 0
    }
}
list1
```

```python

for each in res:
    if each[2]==1:
        if each[3]==0:
            list1['men']['asymptomatic'] += 1
        elif each[3]==1:
            list1['men']['atypical angina'] += 1
        elif each[3]==2:
            list1['men']['non-anginal pain'] += 1
        elif each[3]==3:
            list1['men']['typical angina'] += 1
    if each[2]==0:
        if each[3]==0:
            list1['women']['asymptomatic'] += 1
        elif each[3]==1:
            list1['women']['atypical angina'] += 1
        elif each[3]==2:
            list1['women']['non-anginal pain'] += 1
        elif each[3]==3:
            list1['women']['typical angina'] += 1
list1
```
